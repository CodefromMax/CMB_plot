```python
import healpy as hp
import numpy as np
import matplotlib.pyplot as plt

# Example FITS file path (replace with your actual file path)
fits_file_path = '/Users/maxwang/Documents/projects/CMB_plot/data/planck_simulation/ffp10_newdust_total_030_full_map.fits'

# Read the map
data_map = hp.read_map(fits_file_path)

# Transform the map to alm coefficients
alm = hp.map2alm(data_map)

# Define the range of multipole moments
L = 10
start_l = L
end_l = 2 * L - 1

# Zero out coefficients outside the desired range
for l in range(len(alm)):
    if l < start_l or l > end_l:
        alm[l] = 0

# Transform back to the map
filtered_map = hp.alm2map(alm, nside=hp.get_nside(data_map))

# Plot the filtered map
hp.mollview(filtered_map, title=f'Planck Data Filtered for L in [{start_l}, {end_l}]')
plt.show()
```
